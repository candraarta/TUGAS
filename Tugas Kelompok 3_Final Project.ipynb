```python
import pandas as pd
import matplotlib.pyplot as plt

#luas tanah , lokasi , besar bangunan , kualitas bangunan

train = pd.read_csv("https://raw.githubusercontent.com/imdwipayana/Final-Project/main/House%20Prices/train.csv")
print(train.columns) #get columns in data train
```

```python
print(train['SalePrice'].describe()) #to check sale price 
```

```python
filterTrain = train[['BsmtFinSF1','BedroomAbvGr','LotArea','YearBuilt','OverallQual','OverallCond','YearRemodAdd',
                   'SalePrice']]
print(filterTrain)
```

```python
sortByCheapSale = filterTrain.sort_values(by=['SalePrice'] , ascending = True) #Mengurutkan data dari harga termurah
print(sortByCheapSale.head(5)) 
```

```python
sortByExpensiveSale = filterTrain.sort_values(by = ['SalePrice'] , ascending = False) #mengurutkan data dari termahal
print(sortByExpensiveSale.head(5))
```

```python
sortByLotArea = sortByExpensiveSale.sort_values(by = ['LotArea'] , ascending = False)
lotArea = sortByLotArea.head(5)
```

```python
print(lotArea)
```

```python
plt.plot(lotArea['LotArea'] , color = 'red' , linestyle = 'solid' , label = 'Luas Tanah')
plt.legend(loc='upper right')
plt.show()
plt.plot(lotArea['YearBuilt'], color='blue', linestyle='solid', label='Tahun Dibuat')
plt.plot(lotArea['YearRemodAdd'], color='green', linestyle='dashed', label='Tahun Direnovasi')
plt.legend(loc='upper right')
plt.show()
plt.plot(lotArea['OverallQual'], color='green', linestyle='solid', label='Qualitas')
plt.plot(lotArea['OverallCond'], color='blue', linestyle='dashed', label='Kondisi')
plt.legend(loc='lower right')
plt.show()
```
